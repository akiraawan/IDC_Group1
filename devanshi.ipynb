```python
import os

# Define the path to your main folder
main_folder_path = 'database/testing/'

# Dictionary to store groups and their corresponding patients
groups = {}

# Traverse through each folder in the main folder
try:
    for root, dirs, files in os.walk(main_folder_path):
        for directory in dirs:
            folder_path = os.path.join(root, directory)
            info_file = os.path.join(folder_path, 'Info.cfg')

            # Check if Info.cfg exists in the folder
            if os.path.isfile(info_file):
                try:
                    # Read contents of Info.cfg
                    with open(info_file, 'r') as f:
                        lines = f.readlines()

                        # Extract group information
                        group = None
                        for line in lines:
                            if line.startswith('Group:'):
                                group = line.split(':')[1].strip()
                                break

                        # Add patient to the corresponding group
                        if group:
                            if group not in groups:
                                groups[group] = []
                            groups[group].append(directory)
                except Exception as e:
                    print(f"Error processing {info_file}: {str(e)}")
        
except Exception as e:
    print(f"Error: {e}")
    
# Print groups and their corresponding patients
for group, patients in groups.items():
    print(f'Group: {group}')
    print('Patients:')
    for patient in patients:
        print(f'- {patient}')
```

```python
import os
import shutil

# Define the path to your main folder
main_folder_path = 'database/testing/'

# Define the path to the output folder
output_folder_path = 'database/sorted_data/ยง'

# Create the output folder if it doesn't exist
os.makedirs(output_folder_path, exist_ok=True)

# Traverse through each folder in the main folder
try:
    for root, dirs, files in os.walk(main_folder_path):
        for directory in dirs:
            folder_path = os.path.join(root, directory)
            info_file = os.path.join(folder_path, 'Info.cfg')

            # Check if Info.cfg exists in the folder
            if os.path.isfile(info_file):
                try:
                    # Read contents of Info.cfg
                    with open(info_file, 'r') as f:
                        lines = f.readlines()

                        # Extract group information
                        group = None
                        for line in lines:
                            if line.startswith('Group:'):
                                group = line.split(':')[1].strip()
                                break

                        # Create group directory if it doesn't exist
                        group_folder_path = os.path.join(output_folder_path, group)
                        os.makedirs(group_folder_path, exist_ok=True)
                        print(f"Created group directory: {group_folder_path}")

                        # Move patient folder to the corresponding group directory
                        shutil.move(folder_path, group_folder_path)
                        print(f"Moved patient data to: {group_folder_path}")

                except Exception as e:
                    print(f"Error processing {info_file}: {str(e)}")

except Exception as e:
    print(f"Error: {e}")

print("Data sorted successfully.")
```
